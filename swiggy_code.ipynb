# Exploratory Data Analysis [EDA]

#  Number of Restaurants listed in each City

```python
Total_restaruant_city=df.groupby('City')['Restaurant'].count().sort_values(ascending=False)

total=Total_restaruant_city.plot(kind='bar',color='skyblue')
plt.xticks(rotation=0)
plt.title('Total Restaurant in a city')
sns.set(rc={'figure.figsize':(10,5)})
for bars in total.containers:
    total.bar_label(bars)
```

```python
df.groupby(['City'],as_index=False)['Restaurant'].nunique().sort_values(by='Restaurant',ascending=False)
```

# Finding the total order price by each City

```python
Total_order_price=df.groupby('City')['Price'].sum().sort_values(ascending=False)

total=Total_order_price.plot(kind='bar',color='lightgreen')
plt.xticks(rotation=0)
plt.title('Total order price by each City')
for bars in total.containers:
    total.bar_label(bars)
sns.set(rc={'figure.figsize':(15,6)})
```

# Average order amount from each city

```python
average_order_price=df.groupby('City')['Price'].mean().sort_values(ascending=False)

total=average_order_price.plot(kind='bar',color='green')
plt.xticks(rotation=0)
plt.title('Average order price by each City')
for bars in total.containers:
    total.bar_label(bars)
sns.set(rc={'figure.figsize':(12,5)})
```

```python
sns.histplot(data=df ,x='Avg ratings')
```

```python
mean_total_rating=df.groupby(['City'],as_index=False)['Total ratings'].mean().sort_values(by='Total ratings',ascending=False)

total=sns.barplot(x='City',y='Total ratings',data=mean_total_rating)
plt.xticks(rotation=0)
plt.title('Average Total Rating by each City')
for bars in total.containers:
    total.bar_label(bars)
sns.set(rc={'figure.figsize':(12,5)})
```

# Delivery timings in different city for different foods

```python
Min_Delivery_time=df.groupby(['City'],as_index=False)['Delivery time'].min().sort_values(by='Delivery time')

total=sns.barplot(x='City',y='Delivery time',data=Min_Delivery_time)
plt.title('Minimum delivery timings in city')
for bars in total.containers:
    total.bar_label(bars)
sns.set(rc={'figure.figsize':(12,5)})
```

```python
Max_Delivery_time=df.groupby(['City'],as_index=False)['Delivery time'].max().sort_values(by='Delivery time',ascending=False)

total=sns.barplot(x='City',y='Delivery time',data=Max_Delivery_time)
plt.title('Maximum delivery timings in city')
for bars in total.containers:
    total.bar_label(bars)
sns.set(rc={'figure.figsize':(10,5)})
```

# Conclusion 

The current dataset of Swiggy contains data of 9 cities of the country-Mumbai,Bangalore,Kolkata,Chennai,Pune,Delhi,Ahmedabad,Hyderabad,Surat.

* Kolkata, Mumbai and Hyderabad have the highest restaurants listed.

* Average order in states Mumbai,Bangalore, Kolkata, Chennai and Pune are higher when compared to other with average order rate of rs.350 and above.

* Eatables like Ice cream and chinese fastfood take less time to deliver within a time frame of 20 to 25 mins.

* Based on Average rating of all consumers in all city, Chennai Bangalore and Hyderabad Serves better quality food.

* Few highest rated restaurants based on each city are- majorly milkshake and thickshake outlets, icecream,waffles, Pizza, chinese restaurants, Pan Asian restaurants.

* From the current dataset we can say Mumabi and Kolkata is the highest restaurants liste with the good average order and the faster delivery time.
